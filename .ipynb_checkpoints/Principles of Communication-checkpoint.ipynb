This is not a deterministic function of time, so we have to compute a discrete fourier transform of this data. To do that first we have to index it in form 1: value1,2: value2. For this we use the IndexedSequence function.

```python
# Doing fourier transform of the above data

size = len(amplitude)
# we have to make a fourier transform object known as FFT
fft_data = FFT(size)
```

FFT(size) creats a list of tuples (a,b) where a is real part and b is imaginary part of the foutier trasformed value that we want to evaluate of our data. 
Now we will populate this FFT object with our data. then we will transform it and the function will populate the FFT object with the transformed values.

```python
for j in range(size):
    fft_data[j] = amplitude[j]
```

To now calculate the transform we will use the forward_transform method and then the FFT array we created will get converted to its transformed self. The values that would result would be complex and cant be plotted as is

```python
fft_data.forward_transform()
```

In order to now plot the fourier transform we need to

(a) Convert it into its magnitude form so that its real and positive 
(b) Convert horizontal axis to frequency (currently it is integers)
